# 爬取网易云音乐

```python
import requests
from lxml import etree
import simplejson
import re
import operator
from functools import reduce
```

```python
ua = 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'
headers = {
    'User-agent': ua
}

def get_url_html(url):
    with requests.Session() as session:
        response = session.get(url, headers=headers)
        text = response.text
        html = etree.HTML(text)
    return html

def get_album(html):
        album_ids = html.xpath("//ul[@id='m-song-module']/li/p/a/@href")
        album_names = html.xpath("//ul[@id='m-song-module']/li/p/a/text()")
        album_ids = [ids.split('=')[-1] for ids in album_ids]
        return album_ids, album_names

def get_all_song_id(album_ids):
    with requests.Session() as session:
        all_song_ids, all_song_names = [], []
        ind = 0
        for album_id in album_ids:
            one_album_url = "https://music.163.com/album?id="+str(album_id)
            response = session.get(one_album_url, headers=headers)
            text = response.text
            html = etree.HTML(text)
            album_song_ids = html.xpath("//ul[@class='f-hide']/li/a/@href")
            album_song_names = html.xpath("//ul[@class='f-hide']/li/a/text()")
            album_song_ids = [ids.split('=')[-1] for ids in album_song_ids]

            all_song_ids.append(album_song_ids)
            all_song_names.append(album_song_names)
            
            ind = ind+1
            print('albumn', ind, '/', len(album_ids))

    return all_song_ids, all_song_names
```

```python
album_url = "https://music.163.com/artist/album?id=2116&limit=150&offset=0"  # 列出陈奕迅的所有专辑
html_album = get_url_html(album_url)

album_ids, album_names = get_album(html_album)

all_song_ids, all_song_names = get_all_song_id(album_ids)

all_song_ids_2 = reduce(operator.add, all_song_ids)
all_song_names_2 = reduce(operator.add, all_song_names)
```

# 爬取百度百科

```python
from bs4 import BeautifulSoup
import urllib.request
import re
import pandas as pd
songs = pd.read_csv("song.csv")
```

```python
new_song_names = songs['song'].to_list()
```

```python
song_description = []
num_songs = len(new_song_names)
for i in range(0, num_songs):
    song = new_song_names[i]
    progress = '['+ str(i+1) + '/' + str(num_songs) + ']'
    
    print(progress, "----- 正在爬取歌曲:", song)
    song_str = urllib.parse.quote(song)
    url = "https://baike.baidu.com/item/" + song_str
    response = urllib.request.urlopen(url)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    
    try:
        description = soup.find(attrs={"name":"description"})['content']
        brief = soup.find('div',{'class':'lemma-desc'}).text
        name = soup.find('h1').text
        if '陈奕迅' in brief and '歌曲' in brief:
            print(progress, name, brief, description)
            song_description.append([name, brief, description])
        else:
            ul_data = soup.find_all('ul',{'class':'polysemantList-wrapper cmn-clearfix'})
            if ul_data == []:
                print(progress, '获取失败:', song)
            else:
                print(progress, '多义词:', song)
                for ul in ul_data:
                    tags = ul.find_all(attrs={"title": re.compile('.*陈奕迅.*')})
                    try:
                        for tag in tags:
                            new_url = 'https://baike.baidu.com'+ tag['href']
                            new_response = urllib.request.urlopen(new_url)
                            new_html = new_response.read()
                            new_soup = BeautifulSoup(new_html, 'html.parser')
                            new_description = new_soup.find(attrs={"name":"description"})['content']
                            brief = new_soup.find('div',{'class':'lemma-desc'}).text
                            if '歌曲' in brief:
                                name = new_soup.find('h1').text
                                print(progress, name, brief, new_description)
                                song_description.append([name, brief, new_description])
                            else:
                                print(progress, '获取失败:', song)
                    except:
                        print(progress, '获取失败:', song)
    except:
        print(progress, '获取失败:', song)
```

```python
df = pd.DataFrame(columns = ['歌曲名称', '简要描述', '歌曲介绍'], data = song_description)
df
```

```python
df_new = df.drop_duplicates(subset='歌曲名称', keep='first', inplace=False, ignore_index=False)  # 去重
df_new
```

```python
# 删除翻唱的歌曲
df = df_new
df.drop(df[df['歌曲介绍'].str.contains(pat='翻唱',regex=False)==True].index, inplace=True)
df.drop(df[df['歌曲介绍'].str.contains(pat='原唱',regex=False)==True].index, inplace=True)
df.drop(df[df['简要描述'].str.contains(pat='翻唱',regex=False)==True].index, inplace=True)
df = df.reset_index(drop=True)
```

```python
df.to_csv('data/baike.csv')
```
